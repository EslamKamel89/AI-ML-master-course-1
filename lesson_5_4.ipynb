```python
from utils import pr
from utils import measure_time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from typing import cast

from matplotlib.axes import Axes


rng = np.random.default_rng(42)
income = rng.normal(50000, 10000, 200)
credit_score = rng.normal(650 , 70 , 200)
fig , axes = plt.subplots(1,2 , figsize=(10,8))
axes = cast(list[Axes] , axes)
axes[0].hist(income ,10 , histtype='stepfilled' )
axes[0].set_title('income distribution')
axes[1].hist(credit_score , 10 , histtype='stepfilled')
axes[1].set_title('credit score distribution')
fig.tight_layout()
default = (income < 40000).astype(int)
leak_feature = default + rng.normal( 0 , 0.05 , 200)
```

```python
fig , axes = plt.subplots(1,2 , figsize=(10,4))
axes = cast(list[Axes] , axes)
axes[0].scatter(income , default)
axes[0].set_title("Income vs Default")
axes[0].set_xlabel("Income")
axes[0].set_ylabel("Default")
axes[1].scatter(leak_feature , default)
axes[1].set_title("Leak Feature vs Default")
axes[1].set_xlabel("Leak Feature")
axes[1].set_ylabel("Default")
fig.tight_layout()
```
