```python
from utils import pr
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "user_id": ["U1", "U1", "U2", "U2", "U2", "U3"],
        "transaction_date": pd.to_datetime(
            [
                "2023-01-01",
                "2023-01-10",
                "2023-01-05",
                "2023-01-12",
                "2023-02-01",
                "2023-01-07",
            ]
        ),
        "amount": [100, 200, 50, 75, 300, 500],
        "country": ["US", "US", "US", "US", "US", "FR"],
    }
)
pr(df , 'df')
```

```python
pr(df.groupby('user_id'))
pr(df.groupby('user_id')['amount'])
pr(df.groupby('user_id')['amount'].sum())
pr(df.groupby('user_id')['amount'].sum().reset_index())
```

```python
pr(
    df.groupby('user_id').agg(
        amount_sum = pd.NamedAgg('amount' , 'sum') ,
        amount_avg = pd.NamedAgg('amount' , 'mean') , 
        amount_count = pd.NamedAgg('amount' , 'count')
    )
)
```

```python
pr(df.groupby(['user_id' , 'country'])['amount'].sum())
```

```python
df['month']  = df['transaction_date'].dt.to_period('M')
pr(df)
pr(
    df.groupby(['user_id'  , 'month'])['amount'].sum().reset_index()
)
```
