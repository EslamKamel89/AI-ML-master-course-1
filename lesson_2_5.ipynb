```python
import time
import numpy as np
from typing import  TypeVar 
T = TypeVar('T')
def pr(value:T , title:str|None = None)->T:
    if title :
        print(f'+++++++++++++++++ {title} ++++++++++++++++++')
    else :
        print(f'+++++++++++++++++++++++++++++++++++')
        
    print(type(value))
    print(value)
    print('-----')
    print('')
    return value
def measure_time(func):
    def wrapper(*args,  **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        pr(f"{end_time - start_time} seconds" , 'Execution time')
    return wrapper
```

# when randomness becomes a bug
- this happens when you lose the ability to answer this question `did my changes improved the model or did randomness`

```python
losses :list[float]= []
for _ in range(4) : 
    losses.append(random.random())
pr(losses , 'losses')
```

# transition 
- now that we have seen uncontrolled randomness we need a way to make it predictable when needed.
# seeding randomness

```python
random.seed(42)
values = [random.random() for _ in range(5)]
pr(values , 'seeded values')
```

# determinism vs reproducibility
- deterministic code > same input > same output - always
- reproducible stochastic code > same seed > same code > same output.

```python
np.random.rand()
```

```python
np.random.seed(42)
np.random.rand(5)
```

## critical insight
- seeding python does not seed numpy, later seeding numpy don't mean seeding pytorch
