```python
import time
import numpy as np
from typing import  TypeVar 
T = TypeVar('T')
def pr(value:T , title:str|None = None)->T:
    if title :
        print(f'+++++++++++++++++ {title} ++++++++++++++++++')
    else :
        print(f'+++++++++++++++++++++++++++++++++++')
        
    print(type(value))
    print(value)
    print('-----')
    print('')
    return value
def measure_time(func):
    def wrapper(*args,  **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        pr(f"{end_time - start_time} seconds" , 'Execution time')
    return wrapper
```

```python
data = [1,2,3,4,5,6]
pr(data , 'data?')
for x in data : 
    pr(x , 'x')
```

```python
data = ['1', 2 , 3 , 'hello world']
data.append(True)
pr(data , 'lists allow semantic nonsense')
```

```python
a = [1,2,3]
b = [4,5,6]
pr(a+b , 'Arithmetic on lists is not math')
```

```python
arr = np.array([1,2,3,4,5,6])
pr(arr , 'array')
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
pr(a+b , 'Now arithmetic becomes math')
```

```python
@measure_time
def func1():
    total = 0 
    for x in range(10_000_000):
        total += x
    pr(total , 'func1 total')
@measure_time
def func2():
    arr = np.arange(10_000_000)
    total = arr.sum()
    pr(total , 'func2 total')
func1()
func2()
```
