```python
from utils import pr
import numpy as np
import pandas as pd
```

```python
# Main table: users
users = pd.DataFrame(
    {
        "user_id": ["U1", "U2", "U3", "U4"],
        "signup_date": pd.to_datetime(
            ["2023-01-01", "2023-01-05", "2023-01-10", "2023-01-15"]
        ),
        "churned": [False, True, False, True],
    }
)
#  Secondary table: transactions
transactions = pd.DataFrame(
    {
        "user_id": ["U1", "U1", "U2", "U3", "U3", "U3"],
        "transaction_date": pd.to_datetime(
            [
                "2023-01-03",
                "2023-02-10",
                "2023-01-06",
                "2023-01-12",
                "2023-01-20",
                "2023-02-05",
            ]
        ),
        "amount": [100, 500, 50, 200, 300, 400],
    }
)
pr(users, 'users')
pr(transactions , 'transactions')
```

```python
pr(users.merge(transactions, on="user_id", how="inner"), "inner")
pr(users.merge(transactions, on="user_id", how="outer"), "outer")
pr(users.merge(transactions, on="user_id", how="left"), "left")
pr(users.merge(transactions, on="user_id", how="right"), "right")
```

```python
transactions_summary = (
    transactions.groupby("user_id")
    .agg(
        total_spend=pd.NamedAgg("amount", "sum"),
        transactions_count=pd.NamedAgg("amount", "count"),
    )
    .reset_index()
)
pr(transactions_summary, "transactions_summary")
user_with_features = users.merge(transactions_summary, on="user_id", how="left")
pr(user_with_features, "user_with_features")
pr(len(users) )
pr(len(user_with_features))
```
