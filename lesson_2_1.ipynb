```python
import time
from typing import Generic, TypeVar 
T = TypeVar('T')
def pr(value:T , title:str|None = None)->T:
    if title :
        print(f'+++++++++++++++++ {title} ++++++++++++++++++')
    else :
        print(f'+++++++++++++++++++++++++++++++++++')
        
    print(type(value))
    print(value)
    print('-----')
    print('')
    return value
def measure_time(func):
    def wrapper(*args,  **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        pr(f"{end_time - start_time} seconds" , 'Execution time')
    return wrapper
```

```python
@measure_time
def cell_2() :
    total = 0
    for i in range(10_000_000):
        total += i
    pr(total , 'total')
cell_2()
```

```python
import numpy as np 
@measure_time
def cell_3():
    arr = np.arange(10_000_000)
    pr(arr.sum() , 'total')
cell_3()
```
