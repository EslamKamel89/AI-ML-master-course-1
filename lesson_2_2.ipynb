```python
import time
import numpy as np
from typing import Generic, TypeVar 
T = TypeVar('T')
def pr(value:T , title:str|None = None)->T:
    if title :
        print(f'+++++++++++++++++ {title} ++++++++++++++++++')
    else :
        print(f'+++++++++++++++++++++++++++++++++++')
        
    print(type(value))
    print(value)
    print('-----')
    print('')
    return value
def measure_time(func):
    def wrapper(*args,  **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        pr(f"{end_time - start_time} seconds" , 'Execution time')
    return wrapper
```

```python
x = 0.1
pr(x , 'x')
```

```python
total = 0 
for _ in range(10):
    total += 0.1
pr(total , 'sum of 0.1 ten times')
pr(total == 1 , 'Equality becomes a trap')
```

```python
# types that matters
pr(type(1) , 'int')
pr(type(1.0) , 'float')
pr(type(np.float32(1.0)) ,'float32')
```

```python
a = np.float32(0.1)
total = np.float32(0)
for _ in range(10):
    total += a
pr(total , 'float32 accumulation')
```
