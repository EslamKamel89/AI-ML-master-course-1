```python
from typing import cast

from matplotlib.axes import Axes


fig , axes = plt.subplots(2,2 , figsize=(10,8))
axes = cast(list[Axes], axes.flatten())
for ax , y , i in zip(axes , datasets , range(len(axes))):
    ax.hist( y , bins=5,edgecolor='black')
    ax.set_title(f"Histogram: {i}")
    ax.set_xlabel('x')
    ax.set_ylabel('y')

fig.tight_layout()
```
